```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from matplotlib.ticker import MultipleLocator
```

```python
path = Path(r"C:\Users\dontb\01\001\Repos\Suicide-Analysis\Resources\master.csv")
df = pd.read_csv(path)
df = df.drop(["HDI for year"], axis=1)
df.rename(
    columns={
        "country": "Country",
        "year": "Year",
        "sex": "Sex",
        "age": "Age",
        "suicides_no": "Number of Suicides",
        "population": "Population",
        "suicides/100k pop": "Suicides per 100k",
        "country-year": "Country-Year",
        " gdp_for_year ($) ": "GDP for Year ($)",
        "gdp_per_capita ($)": "GDP per Capita ($)",
        "generation": "Generation",
    },
    inplace=True,
)
us_suicides = df[df["Country"] == "United States"]
us_suicides.to_csv(r"C:\Users\dontb\01\001\Repos\Suicide-Analysis\Resources\us_suicides.csv", index=False)
```

```python
plt.subplots(figsize=(18, 9))
sns.boxplot(x="Year", y="Number of Suicides", data=us_suicides)
plt.grid(True, alpha=0.7)
plt.show()
```
