```python
# Combining and cleaning suicide by state by age data 

df1 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_1999.csv").assign(Year="1999")
df2 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2000.csv").assign(Year="2000")
df3 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2001.csv").assign(Year="2001")
df4 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2002.csv").assign(Year="2002")
df5 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2003.csv").assign(Year="2003")
df6 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2004.csv").assign(Year="2004")
df7 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2005.csv").assign(Year="2005")
df8 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2006.csv").assign(Year="2006")
df9 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2007.csv").assign(Year="2007")
df10 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2008.csv").assign(Year="2008")
df11 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2009.csv").assign(Year="2009")
df12 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2010.csv").assign(Year="2010")
df13 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2011.csv").assign(Year="2011")
df14 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2012.csv").assign(Year="2012")
df15 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2013.csv").assign(Year="2013")
df16 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2014.csv").assign(Year="2014")
df17 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2015.csv").assign(Year="2015")
df18 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2016.csv").assign(Year="2016")
df19 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2017.csv").assign(Year="2017")
df20 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2018.csv").assign(Year="2018")
df21 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2019.csv").assign(Year="2019")
df22 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2020.csv").assign(Year="2020")
df23 = pd.read_csv(r"../Resources/Originals/suicide_by_state_by_age_2021.csv").assign(Year="2021")


dfs = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16,
       df17, df18, df19, df20, df21, df22, df23]

for df in dfs:
    df.set_index("Year", inplace=True, drop=True)
    sum_values = df.iloc[1:].sum(numeric_only=True)
    df.iloc[0] = df.iloc[0].fillna(sum_values)
    column_sums = df.iloc[:, :-1].sum(axis=1, numeric_only=True)
    df["Total"].fillna(column_sums, inplace=True)

suicides_by_age = pd.concat(dfs)
suicides_by_age.drop("0 to 17", axis=1, inplace=True)
suicides_by_age.dropna(subset=["Location"], inplace=True)
suicides_by_age.dropna(inplace=True)
suicides_by_age.to_csv("suicides_by_state_by_age_1999-2021.csv")
```

```python
# Extracting US suicides from World suicides

world_suicides = pd.read_csv(r"../Resources/Originals/world_suicides_1985-2016.csv")
world_suicides.drop(["Population", "Country-Year"], axis=1, inplace=True)
us_suicides_8516 = world_suicides[world_suicides["Country"] == "United States"].copy()
us_suicides_8516.to_csv("us_suicides_1985-2016(master).csv")

# Extracting yearly total suicides for the US from 1985-2013

us_suicides_8513 =  us_suicides_8516.drop(["Sex", "Age", "Suicides per 100k", "GDP for Year ($)", "GDP per Capita ($)", "Generation"], axis=1)
us_suicides_8513 = us_suicides_8513.groupby("Year")["Number of Suicides"].sum().reset_index()
us_suicides_8513 = us_suicides_8513[~((us_suicides_8513["Year"] == 2014) | (us_suicides_8513["Year"] == 2015))]

# Extracting US suicide rates from World suicides

us_suicide_rates_8516 = world_suicides[world_suicides["Country"] == "United States"].copy()
us_suicide_rates_8516 = us_suicide_rates_8516.groupby("Year")["Suicides per 100k"].sum().round(2).reset_index()
us_suicide_rates_8516["Country"] = "United States"
us_suicide_rates_8516 = us_suicide_rates_8516[["Country", "Year", "Suicides per 100k"]].reset_index(drop=True)
us_suicide_rates_8516.to_csv("us_suicide_rates_1985-2016.csv", index=False)

# Extracting US suicide rates from World suicides (1999-2016)

excluded_years = [1985, 1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998]
us_suicide_rates_8516 = world_suicides[world_suicides["Country"] == "United States"].copy()
us_suicide_rates_8516 = us_suicide_rates_8516.groupby("Year")["Suicides per 100k"].sum().round(2).reset_index()
us_suicide_rates_8516 = us_suicide_rates_8516[~us_suicide_rates_8516["Year"].isin(excluded_years)]
us_suicide_rates_8516["Country"] = "United States"
us_suicide_rates_8516 = us_suicide_rates_8516[["Country", "Year", "Suicides per 100k"]].reset_index(drop=True)
us_suicide_rates_8516.to_csv("us_suicide_rates_1999-2016.csv", index=False)
```

```python
# Extracting yearly total suicides for the US from 2014-2021

state_suicides_0521 = pd.read_csv(
    r"../Resources/Originals/state_suicides_2005-2021.csv"
)
state_suicides_0521.drop(["URL", "RATE"], axis=1, inplace=True)
state_suicides_0521["DEATHS"] = state_suicides_0521["DEATHS"].str.replace(",", "")
state_suicides_0521["DEATHS"] = state_suicides_0521["DEATHS"].str.replace('"', "")
state_suicides_0521["DEATHS"] = pd.to_numeric(
    state_suicides_0521["DEATHS"], errors="coerce"
)
us_suicides_0521 = state_suicides_0521.groupby("YEAR")["DEATHS"].sum().reset_index()
us_suicides_0521 = us_suicides_0521[~(us_suicides_0521["YEAR"] == 2005)]
column_mapping = {"YEAR": "Year", "DEATHS": "Number of Suicides"}
us_suicides_0521 = us_suicides_0521.rename(columns=column_mapping)
```

```python
# Concatenating the dataframes to account for missing data

us_suicides_8521 = pd.concat([us_suicides_8513, us_suicides_0521]).reset_index()
us_suicides_8521["Country"] = "United States"
us_suicides_8521 = us_suicides_8521[["Country", "Year", "Number of Suicides"]].reset_index(drop=True)
us_suicides_8521.to_csv("us_suicides_1985-2021.csv", index=False)
```

```python
# Sorting total suicides per 100k by country in descending order

world_suicide_rates_total = world_suicides.copy()

world_suicide_rates_total =  world_suicide_rates_total.drop(["Sex", "Age", "GDP for Year ($)", "GDP per Capita ($)", "Generation"], axis=1)
world_suicide_rates_total = world_suicide_rates_total.groupby("Country")["Suicides per 100k"].sum().reset_index()
world_suicide_rates_total.sort_values("Suicides per 100k", ascending=False, inplace=True)

# Establishing upper and lower quantiles

qlow_total = world_suicide_rates_total["Suicides per 100k"].quantile(0.02)
qhigh_total = world_suicide_rates_total["Suicides per 100k"].quantile(1)
total_filtered = world_suicide_rates_total[(world_suicide_rates_total["Suicides per 100k"] >= qlow_total) & (world_suicide_rates_total["Suicides per 100k"] <= qhigh_total)]

# Extracting top 10 and bottom 10 countries by suicides per 100k

top_10_world_suicides_rates = total_filtered.head(10)
bottom_10_world_suicides_rates = total_filtered.tail(10)
```
